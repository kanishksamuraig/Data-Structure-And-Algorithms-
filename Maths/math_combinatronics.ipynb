## odd or even

#### 1) naive approach

```python
def iseven(number: int):
    if number%2==0:
        return True
    else:
        return False
```

#### 2) efficient(& operation)

```python
def isevenefficient(number: int):
    if (number&1)==0:
        return True
    else:
        return False
```

```python
print(iseven(11)) #nai
print(isevenefficient(14))
```

## Floor and ceil of integer Division

#### built in

```python
import math
def floor(a):
    return math.floor(a)

def ceil(a):
    return math.ceil(a)

print(floor(312/21))
print(ceil(312/21))

```

#### Standard division

```python
import math
def floorstd(a,b):
    return a//b

def ceilstd(a,b):
    return -(-a//b)

print(floorstd(312,21))
print(ceilstd(312,21))
```

## Sum of Natural Numbers

#### Iterative

```python
def findsum(n:int):
    sum=0
    for i in range(n+1):
        sum+=i
    return sum
```

#### Using Formula

```python
def findsumusingformula(n:int):
    return n*(n+1)//2
```

```python
print(findsum(10))
print(findsumusingformula(10))
```

## Sum of the digits

```python
def sumOfDigits(n):
    # code here
    sum=0
    while(n>0):
        sum+=n%10
        n=n//10
    return sum
```

```python
print(sumOfDigits(123))
```

## Decimal to binary conversion

Simple iterative method

```python
def dectobinary(n:int):
    list=[]
    while(n>0):
        list.append(str(n%2))
        n//=2
    return "".join(list)[::-1]
```

```python
print(dectobinary(12))
```

#### Recursive method


```python
def dectobinary(n:int,binarr:list[int]):
    if n==0:
        return

    dectobinary(n//2,binarr)

    binarr.append(str(n%2))
arr=[]
dectobinary(12,arr)
print("".join(arr))

```
