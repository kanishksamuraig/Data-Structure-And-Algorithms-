### dfs and bfs practice

#### Initialize the graph


```python
nodes = int(input("Enter the no of nodes in the graph: "))
graph = [list(map(int,input().split())) for _ in range(nodes)]
```

#### bfs function

```python
print(graph)
from collections import deque
def bfs(graph,nodes,source):
    parent=[-1 for _ in range(nodes)]
    visited=[False for _ in range(nodes)]
    visited[source]=True
    queue=deque()
    queue.append(source)
    while queue:
        node=queue.popleft()
        for neighbour in range(nodes):
            if graph[node][neighbour]==1 and not visited[neighbour]:
                visited[neighbour]=True
                parent[neighbour]=node
                queue.append(neighbour)
    qu=input("Do you want to find the path to a node(yes or no)?:")
    if qu.lower()=="yes":
        dest=int(input("Enter the destination node:"))
        pathfinding(parent,dest)
    print("node parent visited")
    for node,(par,vis) in enumerate(zip(parent,visited)):
        print(f"{node}: {par} {vis}")

def pathfinding(parent,dest):
    path=[]
    node=dest
    path.append(dest)
    while parent[node] != -1:
        node=parent[node]
        path.append(node)
    for i in range(len(path)-1,0,-1):
        print(path[i],"->",end="")
    print(path[0])
```

```python
bfs(graph,nodes,0)
```
